```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

--------------
строим графики значений метрик при изменении параметра units

```python
units = ['1','2', '4', '16', '32', '64', '128']
```

```python
dfs = {'df_' + i: pd.read_csv('/content/drive/MyDrive/diploma/units/1m/metrics_BTC-USD_1m_%s.csv'%i) for i in units}
```

```python
funcs = ['tanh','sigmoid','relu',  'linear', 'loglog', 'cloglog']
names = {j: pd.DataFrame(columns = units) for j in funcs}
```

```python
for j in funcs:
  for i in units:
    names[j][i] = dfs['df_' + i][j]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex = True, figsize=(10, 10))

for j in funcs:
  ax1.plot(units, names[j].iloc[0].to_numpy(), '.-', label = j, alpha = 0.9)
  ax2.plot(units, names[j].iloc[2].to_numpy(), '.-', label = j, alpha = 0.9)
  ax3.plot(units, names[j].iloc[4].to_numpy(), '.-', label = j, alpha = 0.9)

ax1.set(ylabel= 'rmse')
ax2.set(ylabel= 'mae')
ax3.set(ylabel= 'mape', xlabel = 'units')

plt.legend()

plt.savefig('/content/drive/MyDrive/diploma/units/1m/units_1m_train.png', bbox_inches='tight', dpi = 300)
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex = True, figsize=(10, 10))

for j in funcs:
  ax1.plot(units, names[j].iloc[1].to_numpy(), '.-', label = j, alpha = 0.9)
  ax2.plot(units, names[j].iloc[3].to_numpy(), '.-', label = j, alpha = 0.9)
  ax3.plot(units, names[j].iloc[5].to_numpy(), '.-', label = j, alpha = 0.9)

ax1.set(ylabel= 'rmse')
ax2.set(ylabel= 'mae')
ax3.set(ylabel= 'mape', xlabel = 'units')

plt.legend()

plt.savefig('/content/drive/MyDrive/diploma/units/1m/units_1m_test.png', bbox_inches='tight', dpi = 300)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, sharex = True, figsize=(10, 10))

for j in funcs:
  ax1.plot(units, names[j].iloc[2].to_numpy(), '.-', label = j, alpha = 0.9)
  ax2.plot(units, names[j].iloc[3].to_numpy(), '.-', label = j, alpha = 0.9)

ax1.set(ylabel= 'mae (train)')
ax2.set(ylabel= 'mae (test)', xlabel = 'units')

plt.legend()

plt.savefig('/content/drive/MyDrive/diploma/units/1m/units_1m_tt.png', bbox_inches='tight', dpi = 300)
plt.show()
```
